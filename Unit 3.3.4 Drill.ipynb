```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ecommerce = pd.DataFrame()
ecommerce['spending'] = np.random.normal(5, 25, 100)

# Plot a histogram.
plt.hist(ecommerce['spending'])
plt.show()
```

```python
# Plot a histogram for ecommerce random.normal distribution.
plt.hist(ecommerce['spending'], bins=40, color='c')

# Add a vertical line at the mean.
plt.axvline(ecommerce['spending'].mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(ecommerce['spending'].mean() + ecommerce['spending'].std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(ecommerce['spending'].mean()-ecommerce['spending'].std(), color='b', linestyle='dashed', linewidth=2) 
#
# Print the histogram.
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
var1 = np.random.normal(5,.5, 100)
var2 = np.random.normal(10,1, 100)
var3 = var1+var2

mean = np.mean(var3)
sd = np.std(var3)

plt.hist(var3)
plt.axvline(x=mean,color='black')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show()
print('Results will vary depending on random seed.  In general, stats should show some skew.')
```
