Total Farm output from US Farms from 1948 to 2015 

About my dataset

I am using the USDA US Farms indices from 1948-2015 dataset from USDA ERS Department of Agriculture Economic Research Service, this dataset only features indices of the US Farm output, input and total factor productivity from 1948 to 2015. This dataset covers indices of agricultural, and livestock products along with total factor productivity. You can find out more on this dataset here

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

us_farm_dataset = pd.read_csv('usfarm_in_out_tlfactor1948_2015.csv')
```

```python
us_farm_dataset.head()
```

```python
us_farm_dataset_cols = ["Year", "Total_output", "Livestock_and_products_Meat_animals", "Livestock_and_products_Dairy", "Livestock_and_products_Poultry_and_eggs", "Livestock_and_products_All", "Crops_Food_grains",
                       "Crops_Oil_crops", "Crops_Vegetables_and_melons", "Crops_Fruits_and_nuts", "Crops_All"]
us_farm_subset = us_farm_dataset[us_farm_dataset_cols]
```

```python
Year = us_farm_subset.Year
Livestock_and_products_all = us_farm_subset.Livestock_and_products_All
Livestock_and_products_Meat_animals = us_farm_subset.Livestock_and_products_Meat_animals
Livestock_Products_dairy = us_farm_subset.Livestock_and_products_Dairy
Livestock_Products_Poultry = us_farm_subset.Livestock_and_products_Poultry_and_eggs
Crops_Food_grains = us_farm_subset.Crops_Food_grains
Oil_Crops = us_farm_subset.Crops_Oil_crops
Vegetable_and_Melons = us_farm_subset.Crops_Vegetables_and_melons
Fruit_and_nuts_total = us_farm_subset.Crops_Fruits_and_nuts
Crops_total = us_farm_subset.Crops_All
Total_output = us_farm_subset.Total_output
```

```python
us_farm_subset.describe()
```

```python
plt.figure(figsize=(16, 12))
plt.scatter(x=us_farm_dataset['Year'], y=us_farm_dataset['Total_output'])
plt.ylabel('Output based on Indice of 1 in 2005')
plt.xlabel('Years')
plt.title('Total Output Index from US Farms from 1948 - 2015')
plt.show()
```

Analytical Question:

There is breaks in the upward trend of the Market. What is the product makeup of the market that could possibly influence this?

```python
plt.figure(figsize=(16, 12))

plt.subplot(3, 2, 1)
sns.boxplot(y= Fruit_and_nuts_total)
plt.ylim(0, 2)
plt.title('Total Output Index of Fruit and Nut Products 1948-2015')
plt.ylabel('Output based on indice of 1 in 2005')

plt.subplot(3, 2, 2)
sns.boxplot(y= Livestock_and_products_Meat_animals)
plt.ylim(0, 2)
plt.title('Total Output Index of Meat Livestock Products 1948-2015')
plt.ylabel('Output based on indice of 1 in 2005')

plt.subplot(3, 2, 3)
sns.boxplot(y= Livestock_Products_dairy)
plt.ylim(0, 2)
plt.title('Total Output Index of Dairy Products 1948-2015')
plt.ylabel('Output based on indice of 1 in 2005')

plt.subplot(3, 2, 4)
sns.boxplot(y= Vegetable_and_Melons)
plt.ylim(0, 2)
plt.title('Total Output Index of Vegetable and Melon Products 1948-2015')
plt.ylabel('Output based on indice of 1 in 2005')

plt.subplot(3, 2, 5)
sns.boxplot(y= Crops_Food_grains)
plt.ylim(0, 2)
plt.title('Total Output Index of Food Grain Products 1948-2015')
plt.ylabel('Output based on indice of 1 in 2005')

plt.subplot(3, 2, 6)
sns.boxplot(y= Oil_Crops)
plt.ylim(0, 2)
plt.title('Total Output Index of Oil Crop Products 1948-2015')
plt.ylabel('Output based on indice of 1 in 2005')


plt.tight_layout()
plt.show()



```

Above are output indices of the market output. As you can see there is higher indices in the Food Grain,and Livestock Meat Products which most likely influence the market the most. The box plot indicates the center mean of their indices.

```python
sns.factorplot(x= "Year", y= "Total_output", data= us_farm_dataset, kind="bar", 
               palette="BuPu", size=10, aspect=1.5,
               order=np.arange(1948, 2015, 2))
plt.xticks(rotation = 90)
plt.title('Total Output Index of US Farms')
plt.xlabel('Years')
plt.ylabel('Output based on Indice of 1 in 2005')
plt.show()

```

At Second Glance at the upward trend in another plot indicates market increases in 1988.

What products output influenced the market increase in 1982, 1988, 2004 and 2014 of the total output of US farms? 

```python
sns.factorplot(x= "Year", y= "Livestock_and_products_Meat_animals", data= us_farm_dataset, kind="bar", 
               palette="BuPu", size=10, aspect=1.5,
               order=np.arange(1948, 2016, 2))
plt.xticks(rotation = 90)
plt.title('Total Livestock Meat Products Output of US Farms')
plt.xlabel('Years')
plt.ylabel('Output based on Indice of 1 in 2005')
plt.show()
```

```python
sns.factorplot(x= "Year", y= "Fruit_and_nuts_total", data= us_farm_dataset, kind="bar", 
               palette="BuPu", size=10, aspect=1.5, order=np.arange(1948, 2014, 2))
plt.xticks(rotation = 90)
plt.title('Total Food Grain Crop Output of US Farms')
plt.xlabel('Years')
plt.ylabel('Output based on Indice of 1 in 2005')
plt.show()
```

```python
sns.factorplot(x= "Year", y= "Oil_crops", data= us_farm_dataset, kind="bar", 
               palette="BuPu", size=10, aspect=1.5, order=np.arange(1948, 2014, 2))
plt.xticks(rotation = 90)
plt.title('Total Food Grain Crop Output of US Farms')
plt.xlabel('Years')
plt.ylabel('Output based on Indice of 1 in 2005')
plt.show()
```

Based on these two crops that influence the market the most. It appears to come from the Food Grain Crops for the 1982, 1988, amd 2004 increase and the Livestock Meat Products for the 2014 increase in total output.

```python


sns.barplot(x="Year", y="Crops_Food_grains", data=us_farm_subset,
            label="Total Output of Food Grain Crops", palette='BuPu', order=np.arange(1948, 2016))

sns.barplot(x="Year", y='Total_output', data=us_farm_subset,
            label="Total Output of the US farms", palette='GnBu_d', order=np.arange(1948, 2014))

plt.legend(ncol=2, loc="upper left", frameon=True)
plt.title('How much of a difference is the Total output to Food Grains Output')
plt.xlabel("Years")
plt.ylabel("Market Output")
plt.xticks(rotation = 90)
sns.despine(left=False, bottom=True)
plt.show()
```
