```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

welcome_apc = pd.read_csv('WELLCOME_APCspend2013_forThinkful.csv')
```

```python
welcome_apc.head()
```

```python
#word split for cost column for the £ sign to be subtracted from the string. Using strip wipes away leading items to numerical digits. 
#Using split will leave in white space and comma.  
cost_strip = welcome_apc['COST (£) charged to Wellcome (inc VAT when charged)'].str.strip('£$')
```

```python
print(cost_strip)
```

```python
# need to subtract empty space and comma using lambda and also generates things into numerical values
cost_strip = pd.Series(cost_strip)
cost_strip.apply(lambda x:float(x))
```

```python
#now we adding back in the float we just setup. 
welcome_apc['cost_minus']=cost_strip
```

```python
welcome_apc.head()
```

```python
# dropna gets rid of the NaN values in the column
welcome_apc.dropna().groupby(['Journal title','Article title','cost_minus'])
```

```python
welcome_apc.groupby('cost_minus').sum()
```

```python
# check for numerical matches to discover any bad values
print (pd.to_numeric(welcome_apc['cost_minus'], errors='coerce').isnull())
```

```python
#
welcome_apc[~welcome_apc.applymap(np.isreal).all(1)]
```
